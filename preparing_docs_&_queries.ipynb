# 100 dokumentów

```python
import pandas as pd
```

```python
# wczytanie danych
df = pd.read_csv('bbc-news-data.csv', sep='\t' )
```

```python
# sprawdzenie czy kolumny wczytały się poprawnie
print("Dostępne kolumny:", df.columns)
```

```python
# wybór 100 losowych wierszy
subset = df.sample(n=100, random_state=42)
```

```python
# wyciągnięcie kolumny 'content'
texts_only = subset['content']
```

```python
# zapis do pliku tekstowego dokumenty.txt
with open('dokumenty.txt', 'w', encoding='utf-8') as f:
    full_text = '\n\n'.join(texts_only.tolist())
    f.write(full_text)

print(f"Sukces! Zapisano {len(texts_only)} dokumentów do pliku 'dokumenty.txt'.")
```

```python
print("\nPodgląd pierwszego dokumentu:\n", texts_only.iloc[0][:200], "...")
```
